```python
-- 统计每月市场化经营主体数量
SELECT 
    DATE_FORMAT(披露日期, '%Y-%m') AS 月份,
    COUNT(*) AS 市场化经营主体数量
FROM 城投平台市场化经营主体
GROUP BY DATE_FORMAT(披露日期, '%Y-%m')
ORDER BY 月份;

-- 统计每月退出主体数量
SELECT 
    DATE_FORMAT(披露日期, '%Y-%m') AS 月份,
    COUNT(*) AS 退出主体数量
FROM 城投平台退出
GROUP BY DATE_FORMAT(披露日期, '%Y-%m')
ORDER BY 月份;

-- 统计每月两类主体合计数量
SELECT 
    DATE_FORMAT(披露日期, '%Y-%m') AS 月份,
    COUNT(*) AS 合计数量
FROM (
    SELECT 披露日期 FROM 城投平台市场化经营主体
    UNION ALL
    SELECT 披露日期 FROM 城投平台退出
) AS combined
GROUP BY DATE_FORMAT(披露日期, '%Y-%m')
ORDER BY 月份;
```
