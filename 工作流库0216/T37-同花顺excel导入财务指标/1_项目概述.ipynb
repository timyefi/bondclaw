# T37 - 同花顺Excel导入财务指标

## 项目概述

本项目专注于从同花顺Excel文件中提取和处理企业的财务指标数据，包括资产负债表、利润表、现金流量表等关键财务数据，为基本面分析和信用评估提供数据支持。

## 1. 项目背景

### 1.1 财务指标的重要性

1. **信用评估**：财务指标是信用评估的核心依据
2. **基本面分析**：反映企业经营状况和财务健康度
3. **投资决策**：为债券投资提供财务基本面支持
4. **风险识别**：识别财务风险和潜在违约风险

### 1.2 主要财务报表

1. **资产负债表**：反映企业的资产、负债和权益
2. **利润表**：反映企业的收入、成本和利润
3. **现金流量表**：反映企业的现金流入流出
4. **所有者权益变动表**：反映所有者权益的变化

## 2. 项目定位

**同花顺iFinD财务指标批量处理系统**
- 从同花顺API批量提取债券发行人财务指标
- Excel模板生成+数据填充+批量计算
- 数据导入PostgreSQL并转换格式

## 3. 模块结构

### 模块一：同花顺API批量提取
**功能**：从同花顺API提取债券发行人财务指标

### 模块二：批量Excel自动填充内容
**功能**：批量填充Excel日期列（从文件名提取）

### 模块三：Excel数据批量导入到PostgreSQL
**功能**：将Excel数据导入PostgreSQL并转换格式

### 模块四：批量Excel计算及保存为值
**功能**：批量执行Excel计算并保存为值

## 4. 核心数据流

```
数据库查询发行人 -> 同花顺API提取财务指标 -> Excel模板 -> 填充日期 -> Excel计算 -> 导入PostgreSQL -> 格式转换
       ↓                  ↓                   ↓           ↓          ↓           ↓              ↓
    MySQL             THS_BD()            openpyxl    win32com    win32com    pandas       SQL ALTER
```
