```python
#第一步，最好是从网站的svg源码提取，效果最好
svg_content = '''<svg version="1.1" class="highcharts-root" style="font-family: &quot;Lucida Grande&quot;, &quot;Lucida Sans Unicode&quot;, Arial, Helvetica, sans-serif; font-size: 12px;" xmlns="http://www.w3.org/2000/svg" width="854" height="410" viewBox="0 0 854 410"><desc>Created with Highcharts 10.1.0</desc><defs><clipPath id="highcharts-iujbe1r-1-"><rect x="0" y="0" width="793" height="295" fill="none"></rect></clipPath><clipPath id="highcharts-iujbe1r-2-"><rect x="25" y="40" width="793" height="295" fill="none"></rect></clipPath><clipPath id="highcharts-iujbe1r-15-"><rect x="0" y="0" width="793" height="295" fill="none"></rect></clipPath></defs><rect fill="#ffffff" class="highcharts-background" x="0" y="0" width="854" height="410" rx="0" ry="0"></rect><rect fill="none" class="highcharts-plot-background" x="25" y="40" width="793" height="295"></rect><rect fill="none" class="highcharts-plot-border" data-z-index="1" x="25" y="40" width="793" height="295"></rect><g class="highcharts-grid highcharts-yaxis-grid" data-z-index="1"></g><g class="highcharts-grid highcharts-yaxis-grid" data-z-index="1"><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 25 335.5 L 818 335.5" opacity="1"></path><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 25 295.5 L 818 295.5" opacity="1"></path><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 25 244.5 L 818 244.5" opacity="1"></path><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 25 193.5 L 818 193.5" opacity="1"></path><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 25 142.5 L 818 142.5" opacity="1"></path><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 25 91.5 L 818 91.5" opacity="1"></path><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 25 39.5 L 818 39.5" opacity="1"></path></g><g class="highcharts-grid highcharts-xaxis-grid" data-z-index="1"><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 34.5 40 L 34.5 335" opacity="1"></path><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 246.5 40 L 246.5 335" opacity="1"></path><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 458.5 40 L 458.5 335" opacity="1"></path><path fill="none" stroke="#E7E7E7" stroke-width="1" stroke-dasharray="1,1" data-z-index="1" class="highcharts-grid-line" d="M 670.5 40 L 670.5 335" opacity="1"></path></g><g class="highcharts-axis highcharts-yaxis" data-z-index="2"><path fill="none" class="highcharts-axis-line" stroke="#E6E6E6" stroke-width="1" data-z-index="7" d="M 24.5 40 L 24.5 335"></path></g><g class="highcharts-axis highcharts-yaxis" data-z-index="2"><path fill="none" class="highcharts-tick" stroke="#E7E7E7" stroke-width="0.5" d="M 818 335.25 L 828 335.25" opacity="1"></path><path fill="none" class="highcharts-tick" stroke="#E7E7E7" stroke-width="0.5" d="M 818 295.25 L 828 295.25" opacity="1"></path><path fill="none" class="highcharts-tick" stroke="#E7E7E7" stroke-width="0.5" d="M 818 244.25 L 828 244.25" opacity="1"></path><path fill="none" class="highcharts-tick" stroke="#E7E7E7" stroke-width="0.5" d="M 818 193.25 L 828 193.25" opacity="1"></path><path fill="none" class="highcharts-tick" stroke="#E7E7E7" stroke-width="0.5" d="M 818 142.25 L 828 142.25" opacity="1"></path><path fill="none" class="highcharts-tick" stroke="#E7E7E7" stroke-width="0.5" d="M 818 91.25 L 828 91.25" opacity="1"></path><path fill="none" class="highcharts-tick" stroke="#E7E7E7" stroke-width="0.5" d="M 818 39.75 L 828 39.75" opacity="1"></path><path fill="none" class="highcharts-axis-line" stroke="#E6E6E6" stroke-width="1" data-z-index="7" d="M 818.5 40 L 818.5 335"></path></g><g class="highcharts-axis highcharts-xaxis" data-z-index="2"><path fill="none" class="highcharts-axis-line" stroke="#E6E6E6" stroke-width="1" data-z-index="7" d="M 25 335.5 L 818 335.5"></path></g><g class="highcharts-series-group" data-z-index="3"><g class="highcharts-series highcharts-series-0 highcharts-column-series highcharts-dense-data highcharts-tracker" data-z-index="4" opacity="1" transform="translate(25,40) scale(1 1)" clip-path="url(#highcharts-iujbe1r-15-)"><rect x="0" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="0" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="1" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="1" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="2" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="2" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="4" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="4" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="5" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="5" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="6" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="6" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="7" y="81" width="1" height="124" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="7" y="81" width="1" height="124" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="8" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="8" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="10" y="117" width="1" height="88" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="10" y="117" width="1" height="88" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="11" y="77" width="1" height="128" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="11" y="77" width="1" height="128" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="12" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="12" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="13" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="13" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="14" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="14" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="15" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="15" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="17" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="17" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="18" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="18" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="19" y="180" width="1" height="25" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="19" y="180" width="1" height="25" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="20" y="171" width="1" height="34" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="20" y="171" width="1" height="34" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="21" y="195" width="1" height="10" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="21" y="195" width="1" height="10" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="23" y="205" width="1" height="6" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="23" y="205" width="1" height="6" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="24" y="190" width="1" height="15" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="24" y="190" width="1" height="15" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="25" y="201" width="1" height="4" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="25" y="201" width="1" height="4" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="26" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="26" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="27" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="27" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="28" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="28" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="30" y="205" width="1" height="4" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="30" y="205" width="1" height="4" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="31" y="202" width="1" height="3" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="31" y="202" width="1" height="3" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="32" y="186" width="1" height="19" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="32" y="186" width="1" height="19" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="33" y="177" width="1" height="28" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="33" y="177" width="1" height="28" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="34" y="184" width="1" height="21" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="34" y="184" width="1" height="21" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="35" y="194" width="1" height="11" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="35" y="194" width="1" height="11" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="37" y="186" width="1" height="19" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="37" y="186" width="1" height="19" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="38" y="205" width="1" height="4" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="38" y="205" width="1" height="4" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="39" y="205" width="1" height="3" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="39" y="205" width="1" height="3" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="40" y="205" width="1" height="3" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="40" y="205" width="1" height="3" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="41" y="205" width="1" height="39" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="41" y="205" width="1" height="39" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="43" y="198" width="1" height="7" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="43" y="198" width="1" height="7" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="44" y="190" width="1" height="15" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="44" y="190" width="1" height="15" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="45" y="175" width="1" height="30" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="45" y="175" width="1" height="30" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="46" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="46" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="47" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="47" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="48" y="139" width="1" height="66" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="48" y="139" width="1" height="66" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="50" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="50" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="51" y="95" width="1" height="110" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="51" y="95" width="1" height="110" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="52" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="52" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="53" y="64" width="1" height="141" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="53" y="64" width="1" height="141" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="54" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="54" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="56" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="56" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="57" y="67" width="1" height="138" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="57" y="67" width="1" height="138" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="58" y="55" width="1" height="150" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="58" y="55" width="1" height="150" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="59" y="57" width="1" height="148" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="59" y="57" width="1" height="148" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="60" y="90" width="1" height="115" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="60" y="90" width="1" height="115" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="61" y="66" width="1" height="139" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="61" y="66" width="1" height="139" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="63" y="69" width="1" height="136" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="63" y="69" width="1" height="136" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="64" y="68" width="1" height="137" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="64" y="68" width="1" height="137" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="65" y="45" width="1" height="160" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="65" y="45" width="1" height="160" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="66" y="49" width="1" height="156" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="66" y="49" width="1" height="156" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="67" y="56" width="1" height="149" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="67" y="56" width="1" height="149" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="68" y="19" width="1" height="186" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="68" y="19" width="1" height="186" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="70" y="85" width="1" height="120" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="70" y="85" width="1" height="120" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="71" y="64" width="1" height="141" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="71" y="64" width="1" height="141" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="72" y="97" width="1" height="108" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="72" y="97" width="1" height="108" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="73" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="73" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="74" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="74" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="76" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="76" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="77" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="77" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="78" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="78" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="79" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="79" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="80" y="80" width="1" height="125" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="80" y="80" width="1" height="125" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="81" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="81" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="83" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="83" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="84" y="76" width="1" height="129" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="84" y="76" width="1" height="129" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="85" y="67" width="1" height="138" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="85" y="67" width="1" height="138" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="86" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="86" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="87" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="87" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="89" y="176" width="1" height="29" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="89" y="176" width="1" height="29" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="90" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="90" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="91" y="174" width="1" height="31" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="91" y="174" width="1" height="31" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="92" y="195" width="1" height="10" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="92" y="195" width="1" height="10" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="93" y="183" width="1" height="22" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="93" y="183" width="1" height="22" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="94" y="179" width="1" height="26" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="94" y="179" width="1" height="26" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="96" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="96" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="97" y="205" width="1" height="40" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="97" y="205" width="1" height="40" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="98" y="157" width="1" height="48" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="98" y="157" width="1" height="48" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="99" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="99" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="100" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="100" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="101" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="101" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="103" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="103" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="104" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="104" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="105" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="105" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="106" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="106" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="107" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="107" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="109" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="109" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="110" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="110" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="111" y="77" width="1" height="128" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="111" y="77" width="1" height="128" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="112" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="112" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="113" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="113" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="114" y="91" width="1" height="114" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="114" y="91" width="1" height="114" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="116" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="116" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="117" y="40" width="1" height="165" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="117" y="40" width="1" height="165" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="118" y="52" width="1" height="153" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="118" y="52" width="1" height="153" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="119" y="43" width="1" height="162" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="119" y="43" width="1" height="162" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="120" y="50" width="1" height="155" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="120" y="50" width="1" height="155" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="121" y="71" width="1" height="134" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="121" y="71" width="1" height="134" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="123" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="123" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="124" y="49" width="1" height="156" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="124" y="49" width="1" height="156" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="125" y="73" width="1" height="132" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="125" y="73" width="1" height="132" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="126" y="73" width="1" height="132" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="126" y="73" width="1" height="132" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="127" y="84" width="1" height="121" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="127" y="84" width="1" height="121" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="129" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="129" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="130" y="88" width="1" height="117" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="130" y="88" width="1" height="117" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="131" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="131" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="132" y="76" width="1" height="129" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="132" y="76" width="1" height="129" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="133" y="78" width="1" height="127" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="133" y="78" width="1" height="127" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="134" y="95" width="1" height="110" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="134" y="95" width="1" height="110" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="136" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="136" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="137" y="88" width="1" height="117" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="137" y="88" width="1" height="117" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="138" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="138" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="139" y="103" width="1" height="102" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="139" y="103" width="1" height="102" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="140" y="136" width="1" height="69" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="140" y="136" width="1" height="69" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="142" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="142" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="143" y="117" width="1" height="88" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="143" y="117" width="1" height="88" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="144" y="177" width="1" height="28" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="144" y="177" width="1" height="28" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="145" y="205" width="1" height="10" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="145" y="205" width="1" height="10" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="146" y="183" width="1" height="22" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="146" y="183" width="1" height="22" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="147" y="165" width="1" height="40" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="147" y="165" width="1" height="40" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="149" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="149" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="150" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="150" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="151" y="137" width="1" height="68" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="151" y="137" width="1" height="68" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="152" y="156" width="1" height="49" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="152" y="156" width="1" height="49" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="153" y="187" width="1" height="18" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="153" y="187" width="1" height="18" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="154" y="141" width="1" height="64" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="154" y="141" width="1" height="64" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="156" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="156" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="157" y="107" width="1" height="98" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="157" y="107" width="1" height="98" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="158" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="158" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="159" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="159" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="160" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="160" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="162" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="162" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="163" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="163" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="164" y="205" width="1" height="5" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="164" y="205" width="1" height="5" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="165" y="140" width="1" height="65" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="165" y="140" width="1" height="65" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="166" y="199" width="1" height="6" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="166" y="199" width="1" height="6" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="167" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="167" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="169" y="163" width="1" height="42" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="169" y="163" width="1" height="42" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="170" y="192" width="1" height="13" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="170" y="192" width="1" height="13" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="171" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="171" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="172" y="143" width="1" height="62" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="172" y="143" width="1" height="62" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="173" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="173" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="175" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="175" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="176" y="101" width="1" height="104" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="176" y="101" width="1" height="104" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="177" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="177" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="178" y="85" width="1" height="120" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="178" y="85" width="1" height="120" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="179" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="179" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="180" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="180" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="182" y="136" width="1" height="69" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="182" y="136" width="1" height="69" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="183" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="183" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="184" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="184" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="185" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="185" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="186" y="55" width="1" height="150" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="186" y="55" width="1" height="150" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="187" y="67" width="1" height="138" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="187" y="67" width="1" height="138" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="189" y="71" width="1" height="134" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="189" y="71" width="1" height="134" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="190" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="190" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="191" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="191" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="192" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="192" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="193" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="193" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="195" y="123" width="1" height="82" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="195" y="123" width="1" height="82" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="196" y="159" width="1" height="46" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="196" y="159" width="1" height="46" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="197" y="180" width="1" height="25" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="197" y="180" width="1" height="25" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="198" y="181" width="1" height="24" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="198" y="181" width="1" height="24" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="199" y="179" width="1" height="26" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="199" y="179" width="1" height="26" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="200" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="200" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="202" y="196" width="1" height="9" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="202" y="196" width="1" height="9" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="203" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="203" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="204" y="184" width="1" height="21" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="204" y="184" width="1" height="21" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="205" y="178" width="1" height="27" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="205" y="178" width="1" height="27" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="206" y="177" width="1" height="28" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="206" y="177" width="1" height="28" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="208" y="190" width="1" height="15" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="208" y="187" width="1" height="18" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="209" y="194" width="1" height="11" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="209" y="194" width="1" height="11" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="210" y="140" width="1" height="65" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="210" y="141" width="1" height="64" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="211" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="211" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="212" y="101" width="1" height="104" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="212" y="104" width="1" height="101" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="213" y="76" width="1" height="129" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="213" y="76" width="1" height="129" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="215" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="215" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="216" y="93" width="1" height="112" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="216" y="91" width="1" height="114" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="217" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="217" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="218" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="218" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="219" y="65" width="1" height="140" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="219" y="66" width="1" height="139" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="220" y="64" width="1" height="141" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="220" y="61" width="1" height="144" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="222" y="43" width="1" height="162" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="222" y="42" width="1" height="163" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="223" y="81" width="1" height="124" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="223" y="84" width="1" height="121" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="224" y="69" width="1" height="136" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="224" y="71" width="1" height="134" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="225" y="48" width="1" height="157" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="225" y="51" width="1" height="154" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="226" y="96" width="1" height="109" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="226" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="228" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="228" y="108" width="1" height="97" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="229" y="152" width="1" height="53" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="229" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="230" y="109" width="1" height="96" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="230" y="104" width="1" height="101" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="231" y="108" width="1" height="97" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="231" y="104" width="1" height="101" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="232" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="232" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="233" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="233" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="235" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="235" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="236" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="236" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="237" y="101" width="1" height="104" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="237" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="238" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="238" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="239" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="239" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="241" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="241" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="242" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="242" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="243" y="151" width="1" height="54" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="243" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="244" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="244" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="245" y="184" width="1" height="21" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="245" y="180" width="1" height="25" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="246" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="246" y="159" width="1" height="46" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="248" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="248" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="249" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="249" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="250" y="101" width="1" height="104" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="250" y="103" width="1" height="102" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="251" y="137" width="1" height="68" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="251" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="252" y="157" width="1" height="48" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="252" y="158" width="1" height="47" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="253" y="191" width="1" height="14" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="253" y="188" width="1" height="17" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="255" y="195" width="1" height="10" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="255" y="196" width="1" height="9" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="256" y="173" width="1" height="32" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="256" y="171" width="1" height="34" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="257" y="168" width="1" height="37" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="257" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="258" y="198" width="1" height="7" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="258" y="194" width="1" height="11" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="259" y="201" width="1" height="4" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="259" y="201" width="1" height="4" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="261" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="261" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="262" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="262" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="263" y="193" width="1" height="12" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="263" y="196" width="1" height="9" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="264" y="198" width="1" height="7" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="264" y="200" width="1" height="5" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="265" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="265" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="266" y="180" width="1" height="25" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="266" y="179" width="1" height="26" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="268" y="189" width="1" height="16" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="268" y="183" width="1" height="22" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="269" y="143" width="1" height="62" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="269" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="270" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="270" y="140" width="1" height="65" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="271" y="175" width="1" height="30" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="271" y="177" width="1" height="28" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="272" y="162" width="1" height="43" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="272" y="160" width="1" height="45" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="273" y="102" width="1" height="103" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="273" y="104" width="1" height="101" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="275" y="137" width="1" height="68" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="275" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="276" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="276" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="277" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="277" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="278" y="165" width="1" height="40" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="278" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="279" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="279" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="281" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="281" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="282" y="205" width="1" height="7" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="282" y="205" width="1" height="3" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="283" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="283" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="284" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="284" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="285" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="285" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="286" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="286" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="288" y="89" width="1" height="116" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="288" y="94" width="1" height="111" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="289" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="289" y="107" width="1" height="98" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="290" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="290" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="291" y="65" width="1" height="140" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="291" y="66" width="1" height="139" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="292" y="62" width="1" height="143" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="292" y="60" width="1" height="145" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="294" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="294" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="295" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="295" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="296" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="296" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="297" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="297" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="298" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="298" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="299" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="299" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="301" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="301" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="302" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="302" y="140" width="1" height="65" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="303" y="200" width="1" height="5" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="303" y="196" width="1" height="9" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="304" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="304" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="305" y="181" width="1" height="24" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="305" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="306" y="202" width="1" height="3" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="306" y="201" width="1" height="4" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="308" y="205" width="1" height="20" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="308" y="205" width="1" height="17" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="309" y="205" width="1" height="36" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="309" y="205" width="1" height="34" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="310" y="199" width="1" height="6" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="310" y="202" width="1" height="3" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="311" y="188" width="1" height="17" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="311" y="189" width="1" height="16" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="312" y="176" width="1" height="29" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="312" y="177" width="1" height="28" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="314" y="191" width="1" height="14" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="314" y="194" width="1" height="11" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="315" y="189" width="1" height="16" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="315" y="192" width="1" height="13" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="316" y="205" width="1" height="38" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="316" y="205" width="1" height="41" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="317" y="202" width="1" height="3" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="317" y="199" width="1" height="6" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="318" y="205" width="1" height="8" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="318" y="205" width="1" height="6" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="319" y="187" width="1" height="18" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="319" y="187" width="1" height="18" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="321" y="205" width="1" height="32" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="321" y="205" width="1" height="28" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="322" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="322" y="168" width="1" height="37" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="323" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="323" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="324" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="324" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="325" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="325" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="327" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="327" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="328" y="105" width="1" height="100" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="328" y="105" width="1" height="100" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="329" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="329" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="330" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="330" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="331" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="331" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="332" y="154" width="1" height="51" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="332" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="334" y="139" width="1" height="66" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="334" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="335" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="335" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="336" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="336" y="104" width="1" height="101" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="337" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="337" y="125" width="1" height="80" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="338" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="338" y="137" width="1" height="68" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="339" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="339" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="341" y="86" width="1" height="119" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="341" y="86" width="1" height="119" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="342" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="342" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="343" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="343" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="344" y="107" width="1" height="98" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="344" y="105" width="1" height="100" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="345" y="117" width="1" height="88" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="345" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="347" y="151" width="1" height="54" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="347" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="348" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="348" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="349" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="349" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="350" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="350" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="351" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="351" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="352" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="352" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="354" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="354" y="136" width="1" height="69" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="355" y="163" width="1" height="42" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="355" y="164" width="1" height="41" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="356" y="205" width="1" height="15" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="356" y="205" width="1" height="11" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="357" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="357" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="358" y="139" width="1" height="66" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="358" y="136" width="1" height="69" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="360" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="360" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="361" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="361" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="362" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="362" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="363" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="363" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="364" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="364" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="365" y="97" width="1" height="108" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="365" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="367" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="367" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="368" y="98" width="1" height="107" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="368" y="95" width="1" height="110" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="369" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="369" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="370" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="370" y="107" width="1" height="98" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="371" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="371" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="372" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="372" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="374" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="374" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="375" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="375" y="94" width="1" height="111" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="376" y="163" width="1" height="42" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="376" y="162" width="1" height="43" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="377" y="102" width="1" height="103" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="377" y="105" width="1" height="100" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="378" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="378" y="157" width="1" height="48" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="380" y="175" width="1" height="30" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="380" y="175" width="1" height="30" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="381" y="163" width="1" height="42" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="381" y="163" width="1" height="42" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="382" y="186" width="1" height="19" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="382" y="183" width="1" height="22" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="383" y="205" width="1" height="28" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="383" y="205" width="1" height="28" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="384" y="162" width="1" height="43" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="384" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="385" y="143" width="1" height="62" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="385" y="141" width="1" height="64" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="387" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="387" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="388" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="388" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="389" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="389" y="152" width="1" height="53" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="390" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="390" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="391" y="180" width="1" height="25" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="391" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="393" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="393" y="143" width="1" height="62" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="394" y="190" width="1" height="15" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="394" y="190" width="1" height="15" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="395" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="395" y="98" width="1" height="107" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="396" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="396" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="397" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="397" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="398" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="398" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="400" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="400" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="401" y="154" width="1" height="51" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="401" y="154" width="1" height="51" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="402" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="402" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="403" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="403" y="137" width="1" height="68" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="404" y="154" width="1" height="51" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="404" y="154" width="1" height="51" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="405" y="156" width="1" height="49" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="405" y="157" width="1" height="48" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="407" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="407" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="408" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="408" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="409" y="157" width="1" height="48" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="409" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="410" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="410" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="411" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="411" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="413" y="143" width="1" height="62" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="413" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="414" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="414" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="415" y="141" width="1" height="64" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="415" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="416" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="416" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="417" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="417" y="137" width="1" height="68" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="418" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="418" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="420" y="141" width="1" height="64" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="420" y="140" width="1" height="65" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="421" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="421" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="422" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="422" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="423" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="423" y="125" width="1" height="80" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="424" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="424" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="425" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="425" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="427" y="159" width="1" height="46" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="427" y="162" width="1" height="43" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="428" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="428" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="429" y="194" width="1" height="11" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="429" y="195" width="1" height="10" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="430" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="430" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="431" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="431" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="433" y="164" width="1" height="41" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="433" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="434" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="434" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="435" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="435" y="123" width="1" height="82" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="436" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="436" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="437" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="437" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="438" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="438" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="440" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="440" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="441" y="137" width="1" height="68" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="441" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="442" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="442" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="443" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="443" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="444" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="444" y="117" width="1" height="88" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="446" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="446" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="447" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="447" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="448" y="123" width="1" height="82" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="448" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="449" y="67" width="1" height="138" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="449" y="67" width="1" height="138" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="450" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="450" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="451" y="151" width="1" height="54" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="451" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="453" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="453" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="454" y="160" width="1" height="45" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="454" y="160" width="1" height="45" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="455" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="455" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="456" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="456" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="457" y="90" width="1" height="115" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="457" y="91" width="1" height="114" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="458" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="458" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="460" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="460" y="143" width="1" height="62" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="461" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="461" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="462" y="184" width="1" height="21" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="462" y="183" width="1" height="22" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="463" y="205" width="1" height="23" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="463" y="205" width="1" height="22" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="464" y="189" width="1" height="16" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="464" y="185" width="1" height="20" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="466" y="197" width="1" height="8" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="466" y="195" width="1" height="10" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="467" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="467" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="468" y="198" width="1" height="7" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="468" y="197" width="1" height="8" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="469" y="202" width="1" height="3" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="469" y="201" width="1" height="4" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="470" y="185" width="1" height="20" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="470" y="186" width="1" height="19" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="471" y="174" width="1" height="31" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="471" y="175" width="1" height="30" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="473" y="180" width="1" height="25" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="473" y="181" width="1" height="24" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="474" y="202" width="1" height="3" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="474" y="202" width="1" height="3" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="475" y="164" width="1" height="41" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="475" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="476" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="476" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="477" y="168" width="1" height="37" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="477" y="168" width="1" height="37" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="479" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="479" y="164" width="1" height="41" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="480" y="154" width="1" height="51" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="480" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="481" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="481" y="162" width="1" height="43" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="482" y="125" width="1" height="80" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="482" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="483" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="483" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="484" y="184" width="1" height="21" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="484" y="185" width="1" height="20" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="486" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="486" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="487" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="487" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="488" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="488" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="489" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="489" y="171" width="1" height="34" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="490" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="490" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="491" y="160" width="1" height="45" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="491" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="493" y="143" width="1" height="62" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="493" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="494" y="156" width="1" height="49" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="494" y="156" width="1" height="49" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="495" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="495" y="171" width="1" height="34" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="496" y="154" width="1" height="51" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="496" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="497" y="159" width="1" height="46" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="497" y="157" width="1" height="48" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="499" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="499" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="500" y="156" width="1" height="49" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="500" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="501" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="501" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="502" y="139" width="1" height="66" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="502" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="503" y="123" width="1" height="82" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="503" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="504" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="504" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="506" y="123" width="1" height="82" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="506" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="507" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="507" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="508" y="107" width="1" height="98" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="508" y="108" width="1" height="97" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="509" y="123" width="1" height="82" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="509" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="510" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="510" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="512" y="108" width="1" height="97" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="512" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="513" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="513" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="514" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="514" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="515" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="515" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="516" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="516" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="517" y="103" width="1" height="102" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="517" y="103" width="1" height="102" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="519" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="519" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="520" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="520" y="123" width="1" height="82" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="521" y="109" width="1" height="96" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="521" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="522" y="90" width="1" height="115" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="522" y="91" width="1" height="114" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="523" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="523" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="524" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="524" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="526" y="157" width="1" height="48" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="526" y="157" width="1" height="48" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="527" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="527" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="528" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="528" y="152" width="1" height="53" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="529" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="529" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="530" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="530" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="532" y="141" width="1" height="64" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="532" y="139" width="1" height="66" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="533" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="533" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="534" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="534" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="535" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="535" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="536" y="136" width="1" height="69" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="536" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="537" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="537" y="137" width="1" height="68" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="539" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="539" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="540" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="540" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="541" y="178" width="1" height="27" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="541" y="178" width="1" height="27" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="542" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="542" y="151" width="1" height="54" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="543" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="543" y="168" width="1" height="37" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="545" y="136" width="1" height="69" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="545" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="546" y="183" width="1" height="22" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="546" y="181" width="1" height="24" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="547" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="547" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="548" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="548" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="549" y="107" width="1" height="98" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="549" y="108" width="1" height="97" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="550" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="550" y="151" width="1" height="54" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="552" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="552" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="553" y="140" width="1" height="65" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="553" y="139" width="1" height="66" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="554" y="135" width="1" height="70" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="554" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="555" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="555" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="556" y="143" width="1" height="62" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="556" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="557" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="557" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="559" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="559" y="180" width="1" height="25" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="560" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="560" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="561" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="561" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="562" y="183" width="1" height="22" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="562" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="563" y="192" width="1" height="13" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="563" y="192" width="1" height="13" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="565" y="151" width="1" height="54" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="565" y="152" width="1" height="53" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="566" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="566" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="567" y="192" width="1" height="13" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="567" y="189" width="1" height="16" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="568" y="175" width="1" height="30" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="568" y="178" width="1" height="27" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="569" y="163" width="1" height="42" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="569" y="165" width="1" height="40" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="570" y="156" width="1" height="49" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="570" y="158" width="1" height="47" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="572" y="205" width="1" height="13" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="572" y="205" width="1" height="15" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="573" y="205" width="1" height="41" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="573" y="205" width="1" height="39" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="574" y="205" width="1" height="68" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="574" y="205" width="1" height="67" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="575" y="205" width="1" height="64" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="575" y="205" width="1" height="64" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="576" y="205" width="1" height="56" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="576" y="205" width="1" height="55" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="577" y="205" width="1" height="59" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="577" y="205" width="1" height="57" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="579" y="205" width="1" height="13" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="579" y="205" width="1" height="13" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="580" y="205" width="1" height="10" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="580" y="205" width="1" height="12" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="581" y="201" width="1" height="4" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="581" y="198" width="1" height="7" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="582" y="200" width="1" height="5" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="582" y="201" width="1" height="4" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="583" y="198" width="1" height="7" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="583" y="198" width="1" height="7" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="585" y="205" width="1" height="5" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="585" y="205" width="1" height="8" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="586" y="178" width="1" height="27" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="586" y="178" width="1" height="27" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="587" y="163" width="1" height="42" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="587" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="588" y="158" width="1" height="47" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="588" y="159" width="1" height="46" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="589" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="589" y="154" width="1" height="51" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="590" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="590" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="592" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="592" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="593" y="168" width="1" height="37" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="593" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="594" y="174" width="1" height="31" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="594" y="175" width="1" height="30" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="595" y="193" width="1" height="12" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="595" y="191" width="1" height="14" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="596" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="596" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="598" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="598" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="599" y="171" width="1" height="34" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="599" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="600" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="600" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="601" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="601" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="602" y="94" width="1" height="111" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="602" y="95" width="1" height="110" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="603" y="104" width="1" height="101" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="603" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="605" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="605" y="156" width="1" height="49" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="606" y="93" width="1" height="112" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="606" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="607" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="607" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="608" y="116" width="1" height="89" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="608" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="609" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="609" y="163" width="1" height="42" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="610" y="143" width="1" height="62" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="610" y="141" width="1" height="64" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="612" y="173" width="1" height="32" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="612" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="613" y="179" width="1" height="26" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="613" y="181" width="1" height="24" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="614" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="614" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="615" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="615" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="616" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="616" y="154" width="1" height="51" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="618" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="618" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="619" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="619" y="151" width="1" height="54" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="620" y="165" width="1" height="40" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="620" y="165" width="1" height="40" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="621" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="621" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="622" y="176" width="1" height="29" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="622" y="174" width="1" height="31" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="623" y="151" width="1" height="54" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="623" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="625" y="199" width="1" height="6" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="625" y="197" width="1" height="8" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="626" y="171" width="1" height="34" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="626" y="168" width="1" height="37" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="627" y="192" width="1" height="13" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="627" y="194" width="1" height="11" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="628" y="197" width="1" height="8" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="628" y="196" width="1" height="9" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="629" y="196" width="1" height="9" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="629" y="195" width="1" height="10" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="631" y="141" width="1" height="64" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="631" y="142" width="1" height="63" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="632" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="632" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="633" y="149" width="1" height="56" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="633" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="634" y="193" width="1" height="12" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="634" y="194" width="1" height="11" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="635" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="635" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="636" y="160" width="1" height="45" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="636" y="159" width="1" height="46" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="638" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="638" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="639" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="639" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="640" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="640" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="641" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="641" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="642" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="642" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="643" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="643" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="645" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="645" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="646" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="646" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="647" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="647" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="648" y="109" width="1" height="96" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="648" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="649" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="649" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="651" y="117" width="1" height="88" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="651" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="652" y="102" width="1" height="103" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="652" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="653" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="653" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="654" y="145" width="1" height="60" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="654" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="655" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="655" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="656" y="134" width="1" height="71" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="656" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="658" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="658" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="659" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="659" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="660" y="140" width="1" height="65" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="660" y="139" width="1" height="66" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="661" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="661" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="662" y="165" width="1" height="40" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="662" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="664" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="664" y="148" width="1" height="57" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="665" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="665" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="666" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="666" y="168" width="1" height="37" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="667" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="667" y="171" width="1" height="34" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="668" y="144" width="1" height="61" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="668" y="147" width="1" height="58" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="669" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="669" y="108" width="1" height="97" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="671" y="184" width="1" height="21" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="671" y="174" width="1" height="31" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="672" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="672" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="673" y="172" width="1" height="33" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="673" y="171" width="1" height="34" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="674" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="674" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="675" y="187" width="1" height="18" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="675" y="187" width="1" height="18" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="676" y="164" width="1" height="41" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="676" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="678" y="163" width="1" height="42" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="678" y="164" width="1" height="41" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="679" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="679" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="680" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="680" y="169" width="1" height="36" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="681" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="681" y="155" width="1" height="50" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="682" y="138" width="1" height="67" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="682" y="140" width="1" height="65" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="684" y="167" width="1" height="38" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="684" y="165" width="1" height="40" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="685" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="685" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="686" y="131" width="1" height="74" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="686" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="687" y="91" width="1" height="114" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="687" y="91" width="1" height="114" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="688" y="123" width="1" height="82" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="688" y="126" width="1" height="79" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="689" y="123" width="1" height="82" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="689" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="691" y="85" width="1" height="120" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="691" y="88" width="1" height="117" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="692" y="76" width="1" height="129" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="692" y="77" width="1" height="128" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="693" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="693" y="83" width="1" height="122" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="694" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="694" y="97" width="1" height="108" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="695" y="61" width="1" height="144" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="695" y="61" width="1" height="144" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="697" y="61" width="1" height="144" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="697" y="62" width="1" height="143" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="698" y="60" width="1" height="145" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="698" y="56" width="1" height="149" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="699" y="60" width="1" height="145" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="699" y="53" width="1" height="152" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="700" y="64" width="1" height="141" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="700" y="63" width="1" height="142" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="701" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="701" y="75" width="1" height="130" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="702" y="70" width="1" height="135" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="702" y="75" width="1" height="130" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="704" y="66" width="1" height="139" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="704" y="67" width="1" height="138" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="705" y="75" width="1" height="130" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="705" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="706" y="95" width="1" height="110" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="706" y="97" width="1" height="108" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="707" y="122" width="1" height="83" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="707" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="708" y="76" width="1" height="129" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="708" y="75" width="1" height="130" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="709" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="709" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="711" y="99" width="1" height="106" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="711" y="100" width="1" height="105" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="712" y="105" width="1" height="100" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="712" y="100" width="1" height="105" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="713" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="713" y="103" width="1" height="102" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="714" y="71" width="1" height="134" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="714" y="67" width="1" height="138" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="715" y="84" width="1" height="121" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="715" y="89" width="1" height="116" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="717" y="88" width="1" height="117" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="717" y="94" width="1" height="111" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="718" y="93" width="1" height="112" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="718" y="94" width="1" height="111" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="719" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="719" y="130" width="1" height="75" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="720" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="720" y="93" width="1" height="112" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="721" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="721" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="722" y="104" width="1" height="101" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="722" y="101" width="1" height="104" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="724" y="71" width="1" height="134" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="724" y="76" width="1" height="129" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="725" y="120" width="1" height="85" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="725" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="726" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="726" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="727" y="80" width="1" height="125" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="727" y="70" width="1" height="135" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="728" y="127" width="1" height="78" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="728" y="129" width="1" height="76" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="730" y="110" width="1" height="95" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="730" y="117" width="1" height="88" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="731" y="85" width="1" height="120" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="731" y="91" width="1" height="114" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="732" y="96" width="1" height="109" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="732" y="96" width="1" height="109" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="733" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="733" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="734" y="160" width="1" height="45" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="734" y="161" width="1" height="44" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="735" y="150" width="1" height="55" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="735" y="153" width="1" height="52" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="737" y="109" width="1" height="96" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="737" y="106" width="1" height="99" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="738" y="108" width="1" height="97" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="738" y="113" width="1" height="92" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="739" y="121" width="1" height="84" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="739" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="740" y="92" width="1" height="113" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="740" y="88" width="1" height="117" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="741" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="741" y="66" width="1" height="139" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="742" y="112" width="1" height="93" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="742" y="114" width="1" height="91" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="744" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="744" y="124" width="1" height="81" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="745" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="745" y="79" width="1" height="126" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="746" y="107" width="1" height="98" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="746" y="107" width="1" height="98" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="747" y="82" width="1" height="123" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="747" y="80" width="1" height="125" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="748" y="71" width="1" height="134" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="748" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="750" y="69" width="1" height="136" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="750" y="72" width="1" height="133" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="751" y="58" width="1" height="147" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="751" y="54" width="1" height="151" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="752" y="55" width="1" height="150" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="752" y="61" width="1" height="144" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="753" y="82" width="1" height="123" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="753" y="76" width="1" height="129" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="754" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="754" y="115" width="1" height="90" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="755" y="87" width="1" height="118" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="755" y="77" width="1" height="128" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="757" y="94" width="1" height="111" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="758" y="139" width="1" height="66" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="759" y="118" width="1" height="87" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="760" y="128" width="1" height="77" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="761" y="111" width="1" height="94" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="762" y="133" width="1" height="72" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="764" y="173" width="1" height="32" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="765" y="159" width="1" height="46" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="766" y="195" width="1" height="10" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="767" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="768" y="184" width="1" height="21" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="770" y="180" width="1" height="25" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="771" y="183" width="1" height="22" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="772" y="146" width="1" height="59" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="773" y="170" width="1" height="35" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="774" y="174" width="1" height="31" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="775" y="205" width="1" height="3" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="777" y="205" width="1" height="15" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="778" y="197" width="1" height="8" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="779" y="166" width="1" height="39" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="780" y="186" width="1" height="19" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="781" y="205" width="1" height="8" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="783" y="182" width="1" height="23" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="784" y="205" width="1" height="16" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="785" y="205" width="1" height="12" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="786" y="205" width="1" height="27" fill="#f5deb3" opacity="1" class="highcharts-point highcharts-negative"></rect><rect x="787" y="178" width="1" height="27" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="788" y="159" width="1" height="46" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="790" y="119" width="1" height="86" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="791" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect><rect x="792" y="132" width="1" height="73" fill="#4f81bd" opacity="1" class="highcharts-point"></rect></g><g class="highcharts-markers highcharts-series-0 highcharts-column-series highcharts-tracker" data-z-index="4" opacity="1" transform="translate(25,40) scale(1 1)" clip-path="none"></g></g><g class="highcharts-exporting-group" data-z-index="3"><g class="highcharts-no-tooltip highcharts-button highcharts-contextbutton" stroke-linecap="round" style="cursor: pointer;" transform="translate(814,0)"><title>Chart context menu</title><rect fill="#ffffff" class="highcharts-button-box" x="0.5" y="0.5" width="30" height="30" rx="2" ry="2" stroke="none" stroke-width="1"></rect><path fill="#666666" d="M 6 6.5 L 20 6.5 M 6 11.5 L 20 11.5 M 6 16.5 L 20 16.5" class="highcharts-button-symbol" data-z-index="1" stroke="#666666" stroke-width="3"></path><text x="0" data-z-index="1" y="19.5" style="color: rgb(51, 51, 51); font-weight: normal; fill: rgb(51, 51, 51);"></text></g></g><text x="427" text-anchor="middle" class="highcharts-title" data-z-index="4" style="color: rgb(51, 51, 51); font-size: 16px; fill: rgb(51, 51, 51);" y="22"></text><text x="427" text-anchor="middle" class="highcharts-subtitle" data-z-index="4" style="color: rgb(102, 102, 102); fill: rgb(102, 102, 102);" y="24"></text><text x="10" text-anchor="start" class="highcharts-caption" data-z-index="4" style="color: rgb(102, 102, 102); fill: rgb(102, 102, 102);" y="407"></text><g class="highcharts-axis-labels highcharts-yaxis-labels" data-z-index="7"></g><g class="highcharts-axis-labels highcharts-yaxis-labels" data-z-index="7"><text x="0" style="color: rgb(51, 51, 51); cursor: default; font-size: 1em; font-family: Helvetica; fill: rgb(51, 51, 51);" text-anchor="start" transform="translate(0,0)" visibility="hidden"><tspan class="yLabels_">-25</tspan></text><text x="823" style="color: rgb(51, 51, 51); cursor: default; font-size: 1em; font-family: Helvetica; fill: rgb(51, 51, 51);" text-anchor="start" transform="translate(0,0)" y="298" opacity="1"><tspan class="yLabels_">-14</tspan></text><text x="823" style="color: rgb(51, 51, 51); cursor: default; font-size: 1em; font-family: Helvetica; fill: rgb(51, 51, 51);" text-anchor="start" transform="translate(0,0)" y="247" opacity="1"><tspan class="yLabels_">0</tspan></text><text x="823" style="color: rgb(51, 51, 51); cursor: default; font-size: 1em; font-family: Helvetica; fill: rgb(51, 51, 51);" text-anchor="start" transform="translate(0,0)" y="196" opacity="1"><tspan class="yLabels_">14</tspan></text><text x="823" style="color: rgb(51, 51, 51); cursor: default; font-size: 1em; font-family: Helvetica; fill: rgb(51, 51, 51);" text-anchor="start" transform="translate(0,0)" y="145" opacity="1"><tspan class="yLabels_">28</tspan></text><text x="823" style="color: rgb(51, 51, 51); cursor: default; font-size: 1em; font-family: Helvetica; fill: rgb(51, 51, 51);" text-anchor="start" transform="translate(0,0)" y="94" opacity="1"><tspan class="yLabels_">42</tspan></text><text x="0" style="color: rgb(51, 51, 51); cursor: default; font-size: 1em; font-family: Helvetica; fill: rgb(51, 51, 51);" text-anchor="start" transform="translate(0,0)" visibility="hidden"><tspan class="yLabels_">56</tspan></text></g><g class="highcharts-axis-labels highcharts-xaxis-labels" data-z-index="7"></g><g class="highcharts-tooltip" data-z-index="8" opacity="0" y="-9999" visibility="hidden"></g><g class="highcharts-control-points" data-z-index="99" clip-path="url(#highcharts-iujbe1r-2-)"></g></svg>'''

import xml.etree.ElementTree as ET

def extract_unique_coordinates(svg_content):
    # Parse the SVG content
    root = ET.fromstring(svg_content)

    # Namespace dictionary for SVG
    namespaces = {'svg': 'http://www.w3.org/2000/svg'}

    # Find all rect elements (assuming bars are represented as <rect> elements)
    rects = root.findall('.//svg:rect', namespaces)
    filtered_rects = [rect for rect in rects if rect.get('class', None) in ['highcharts-point', 'highcharts-point highcharts-negative']]
    filtered_rects = [rect for rect in filtered_rects if rect.get('fill', None) in ['#4f81bd', '#f5deb3']]
    coordinates = []
    for rect in filtered_rects:
        # print(rect.items())
        # Extract attributes, assuming x and y represent the position of the bar
        x = rect.get('x')
        y = rect.get('y')
        height = float(rect.get('height'))
        class1 = rect.get('class')
        
        # Ensure x and y are not None and then convert to float
        if x is not None and y is not None:
            x = float(x)
            y = float(y)
            
            # Check if the bar is positive or negative
            if class1 == 'highcharts-point':
                # Negative bar, take the left bottom corner
                y_coord = height
            else:
                # Positive bar, take the left top corner
                y_coord = -height
            # Check if this (x, y) coordinate has already been recorded
            if x not in [coord[0] for coord in coordinates]:
                coordinates.append((x, y_coord))
    # Print the count and coordinates
    print(f"There are {len(coordinates)} unique (x, y) values.")
    for x, y in coordinates:
        print(f"Rect coordinate: x = {x}, y = {y}")
    return coordinates

# Example usage
coordinates=extract_unique_coordinates(svg_content)
```

```python
#线图处理参考，第一步先解析svg，分析数据结构，提供给gpt做参考
import xml.etree.ElementTree as ET
import re

def parse_svg(svg_file):
    # 读取SVG文件
    tree = ET.parse(svg_file)
    root = tree.getroot()

    # 定义SVG命名空间
    namespace = {'svg': 'http://www.w3.org/2000/svg'}

    # 查找所有路径元素
    paths = root.findall('.//svg:path', namespace)

    # 用于存储所有路径数据点
    all_points = []

    # 解析路径数据
    for path in paths:
        d = path.attrib.get('d')
        if d:
            # 提取路径中的所有坐标点
            points = re.findall(r'[-+]?\d*\.\d+|[-+]?\d+', d)
            points = list(map(float, points))

            # 分组为(x, y)对
            points = [(points[i], points[i + 1]) for i in range(0, len(points), 2)]
            all_points.extend(points)

    return all_points

# 指定SVG文件路径
svg_file = r'/Users/yetim/Downloads/US_Philly_Fed_CAPEX_Index-2.svg'

# 解析SVG文件并获取路径数据点
points = parse_svg(svg_file)

# 打印结果
# print(f"Total points extracted: {len(points)}")
print(points)


```
