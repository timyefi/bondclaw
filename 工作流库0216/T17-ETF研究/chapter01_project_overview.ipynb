## 8. Quick Start Example

````python
# Quick start example: Load ETF data
try:
    # Initialize data loader
    loader = DataLoader()
    
    # Get latest trading day
    latest_day = loader.query_df("SELECT MAX(DT) as latest FROM fund.marketinfo")
    print(f"Latest trading day: {latest_day['latest'].iloc[0]}")
    
    # Get ETF base info count
    etf_count = loader.query_df("SELECT COUNT(*) as cnt FROM yq.etf_base_info WHERE DT = (SELECT MAX(DT) FROM yq.etf_base_info)")
    print(f"ETF count: {etf_count['cnt'].iloc[0]}")
    
except Exception as e:
    print(f"Database connection error: {e}")
    print("Please check your database configuration and ensure environment variables are set.")
````
