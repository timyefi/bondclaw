## 1. 项目简介

机构行为监测项目是一个用于监测和分析债券市场机构投资者交易行为的系统。通过采集机构交易数据，进行行为模式分析、多维度可视化监测和风险预警。

### 1.1 主要功能

- **机构交易数据采集**: 从Excel文件和数据库同步现券成交分机构统计数据
- **行为模式分析**: 分析机构期限偏好、券种偏好、交易热点等
- **多维度可视化监测**: 提供14种专业监测图表
- **风险预警**: 通过市场核心行为指数、风险定价等指标进行预警

### 1.2 应用场景

- 机构行为研究
- 市场情绪分析
- 投资策略参考

## 8. 核心算法概述

### 8.1 机构行为指标

1. **RFY Premium (评级基准收益率溢价)**
   - 衡量交易收益率相对评级的偏离程度
   - 计算方式: 交易收益率 - 评级基准收益率

2. **Herding Index (羊群效应指数)**
   - 基于LSV模型计算市场羊群效应
   - H = |p_i - p| - AF (调整因子)

3. **Duration of Net Flow (久期净流向)**
   - 计算买入/卖出交易的加权久期
   - 反映机构久期配置倾向

### 8.2 风险定价指标

1. **期限利差**: 10Y国债 - 1Y国债
2. **信用利差**: AA+ 3Y中票 - 3Y国债

### 8.3 资金面指标

1. **回购余额**: 银行间 + 交易所
2. **回购利率**: DR007, R007
